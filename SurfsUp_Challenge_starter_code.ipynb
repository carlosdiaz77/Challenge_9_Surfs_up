```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

from sqlalchemy import extract  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python


results_jun=session.query(Measurement.date, Measurement.tobs).filter(extract('year', Measurement.date)>1900).filter(extract('month', Measurement.date)==6).all()
```

```python
#  3. Convert the June temperatures to a list.
print(results_jun)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
temps_df_jun = pd.DataFrame(results_jun, columns=['date','temperature'])
temps_df_jun
```

# Summary Statistics for June

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
temps_df_jun.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
results_dec=session.query(Measurement.date, Measurement.tobs).filter(extract('year', Measurement.date)>1900).filter(extract('month', Measurement.date)==12).all()
```

```python
# 7. Convert the December temperatures to a list.
print(results_dec)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
temps_df_dec = pd.DataFrame(results_dec, columns=['date','temperature'])
temps_df_dec
```

# Summary Statistics for December

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
temps_df_dec.describe()
```

# Extra code for challenge summary points

```python
jun_series = temps_df_jun.groupby(["date"]).mean()['temperature']

x_labels = ["temperature"]
fig, ax = plt.subplots()
ax.boxplot(jun_series, labels=x_labels)


# Add the title, y-axis label and grid.
ax.set_title('Tempeture in JUN')
ax.set_ylabel('Farenheigt')
ax.set_yticks(np.arange(60, 85, step=2.0))

ax.grid()
plt.show()
```

```python
dec_series = temps_df_dec.groupby(["date"]).mean()['temperature']

x_labels = ["temperature"]
fig, ax = plt.subplots()
ax.boxplot(dec_series, labels=x_labels)


# Add the title, y-axis label and grid.
ax.set_title('Tempeture in DEC')
ax.set_ylabel('Farenheigt')
ax.set_yticks(np.arange(60, 85, step=2.0))

ax.grid()
plt.show()
```
